```python
from flask import Flask, request
import numpy as np
import pickle
import pandas as pd
import flasgger
from flasgger import Swagger

app=Flask(__name__)
Swagger(app)

pickle_in = open("/Users/user/Desktop/DS/python/zomato/ETree.pkl","rb")
ETree=pickle.load(pickle_in)

@app.route('/')
def welcome():
    return "Welcome All"


@app.route('/predict',methods=["Get"])
def zomato_rating():
    """Let's Authenticate the rating of resturant 
    This is using docstrings for specifications.
    ---
    parameters:  
      - name: online_order
        in: query
        type: number
        required: true
      - name: book_table
        in: query
        type: number
        required: true
      - name: votes
        in: query
        type: number
        required: true
      - name: location
        in: query
        type: number
        required: true
      - name: rest_type
        in: query
        type: number
        required: true
      - name: cuisines
        in: query
        type: number
        required: true
      - name: cost
        in: query
        type: number
        required: true
      - name: menu_item
        in: query
        type: number
        required: true
    responses:
        200:
            description: The output values
        
    """
    
    online_order=request.args.get("online_order")
    book_table=request.args.get("book_table")
    votes=request.args.get("votes")
    location=request.args.get("location")
    rest_type=request.args.get("rest_type")
    cuisines=request.args.get("cuisines")
    cost=request.args.get("cost")
    menu_item=request.args.get("menu_item")
    prediction=ETree.predict([[online_order,book_table,votes,location,rest_type,cuisines,cost,menu_item]])
    return "Hello The answer is"+str(prediction)


@app.route('/chg',methods=['Post'])
def zomato_rating_prediction():
    """Let's Authenticate the rating of resturant
    This is using docstrings for specifications.
    ---
    parameters:
      - name: file
        in: formData
        type: file
        required: true
      
    responses:
        200:
            description: The output values
        
    """
    
    df_test=pd.read_csv(request.files.get("file"))
    print(df_test.head())
    prediction=ETree.predict(df_test)
    return str(list(prediction))


@app.route('/index')
def a():
    return "index"
    

if __name__=='__main__':
    app.run()
```
